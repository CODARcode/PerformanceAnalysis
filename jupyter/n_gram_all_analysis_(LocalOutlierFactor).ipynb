```python
import pickle
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import glob
from sklearn.neighbors import LocalOutlierFactor
```

```python
os.getcwd()
os.chdir("../data")
os.getcwd()

df_lst = []

for file in glob.glob("ee*.json.df"):
    print("Processing :" + file)

    with open(file, 'rb') as handle:
        ldf = pd.read_pickle(handle)
        df_lst.append(ldf)

print(len(df_lst))
```

```python
df = pd.concat(df_lst)
```

```python
df = df.sort_values(['time_by_lasttime'], ascending=[True])
```

```python
import pandas as pd
sf = df.kl.value_counts()
freq = pd.DataFrame({'n_gram':sf.index, 'numberofcalls':sf.values})

# Define max and min threshold
min = 10000
max = float("inf")

subset = freq[(freq['numberofcalls'] > min) & (freq['numberofcalls'] < max)]
subset = subset.sample(frac=0.4)
subset
```

```python
def perform_localOutlierFactor(df, n_neighbors=10, contamination=0.001):
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn.neighbors import LocalOutlierFactor
    
    fig = matplotlib.pyplot.gcf()
    fig.set_size_inches(15.5, 10.5)
    
    f_name = df.iloc[0]['kl']    
    
    clf = LocalOutlierFactor(n_neighbors=n_neighbors, # broader estimation 
                             algorithm="auto", leaf_size=30, metric="minkowski", p=2, metric_params=None, contamination=contamination, n_jobs=-1)
    
    y_pred = clf.fit_predict(df[['time_by_lasttime','time_diff']])
    
    minx=df['time_by_lasttime'].min()
    maxx=df['time_by_lasttime'].max()
    miny=df['time_diff'].min()
    maxy=df['time_diff'].max()
    
    plt.title("LocalOutlierFactor; function: "+str(f_name) + "K=" + str(n_neighbors))
        
    # plot outliers in red and normals in white

    # Method 2
    df_outliers = df[ y_pred ==-1 ]
    df_normals = df[ y_pred ==+1 ]
    
    g = plt.scatter(df_normals['time_by_lasttime'],  df_normals['time_diff'], c='green', s=25, edgecolor='k')
    h = plt.scatter(df_outliers['time_by_lasttime'], df_outliers['time_diff'], c='red', s=25, edgecolor='k')
    
    plt.axis('tight')
    plt.xlim((minx, maxx))
    plt.ylim((miny, maxy))
    plt.legend([g, h],["Normal","Outliers"],loc="upper right")
    plt.show()
    
    return df_outliers, df_normals
```

```python
sns.set_context("notebook", font_scale=1.2, rc={"lines.linewidth": 2.5})

df_s1_outliers = []
df_s1_normals = []
df_s2_outliers = []
df_s2_normals = []
for i in range(len(subset)):
    pick_func = freq.iloc[i]['n_gram']
    count     = freq.iloc[i]['numberofcalls']
    lbl       = str(pick_func) +'_count_'+ str(count)
    
    ##############
    # df[df['kl'] == pick_func].plot(x='time_by_lasttime', y='time_diff', kind = 'scatter', title=lbl )
    # plt.show()
    ##############
    
    item_s1_outliers, item_s1_normals = perform_localOutlierFactor(df[df['kl'] == pick_func], 1024, float(10 / count))
    item_s2_outliers, item_s2_normals = perform_localOutlierFactor(df[df['kl'] == pick_func], 10, float(10 / count))
    
    df_s1_outliers.append(item_s1_outliers)
    df_s1_normals.append(item_s1_normals)
    df_s2_outliers.append(item_s2_outliers)
    df_s2_normals.append(item_s2_normals)
```

```python
def convert_df2csv(outliers_list, normals_list, k):
    if len(outliers_list) != len(normals_list):
        raise RuntimeError("outlier list number does not match with normal list number")
        
    num = len(outliers_list)
    for outliers in outliers_list:
        outliers['class'] = 1
    for normals in normals_list:
        normals['class'] = 0
    
    for i in range(num):
        index = outliers_list[i]['kl'].index[0]
        filename = outliers_list[i]['kl'][index] + '_K=' + str(k) + '.csv'
        merged = pd.concat([normals_list[i], outliers_list[i]])
        merged.to_csv(filename)
        
convert_df2csv(df_s1_outliers, df_s1_normals, 1024)
convert_df2csv(df_s2_outliers, df_s2_normals, 10)
```
